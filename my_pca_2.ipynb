```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
```

```python
df=pd.read_csv('wine.csv')
```

```python
df.head()
```

```python
df
```

```python
X=df.iloc[:,1:13].values
```

```python
X
```

```python
y=df.iloc[:,0].values
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
X_train
```

```python
y_test
```

```python
# performing preprocessing part 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 
  
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test) 
```

```python
#preprocessing fitting the standard scale
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
X_train=pca.fit_transform(X_train)
X_test=pca.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
#variance 
```

```python
explained_variance = pca.explained_variance_ratio_ 
```

```python
explained_variance
```

```python
#apply logistic regression to find results 
```

```python
from sklearn.linear_model import LogisticRegression

lg_model=LogisticRegression()

lg_model.fit(X_train,y_train)
```

```python
lg_predictions= lg_model.predict(X_test)
lg_predictions
```

```python
y_test.shape
lg_predictions.shape
```

```python
print(confusion_matrix(y_test,lg_predictions))
print(classification_report(y_test,lg_predictions))
print(lg_model.score(X_test,y_test))

        
```

```python
#accuracy
count=0
for i in range(54):
    if y_test[i]==lg_predictions[i]:
        count=count+1
    else:
        count=count
print(count/54)
```

```python
#further are just to plot the train and test data differently
```

```python
# Predicting the training set 
# result through scatter plot  
from matplotlib.colors import ListedColormap 
  
X_set, y_set = X_train, y_train 
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                     stop = X_set[:, 0].max() + 1, step = 0.01), 
                     np.arange(start = X_set[:, 1].min() - 1, 
                     stop = X_set[:, 1].max() + 1, step = 0.01)) 
  
plt.contourf(X1, X2, lg_model.predict(np.array([X1.ravel(), 
             X2.ravel()]).T).reshape(X1.shape), alpha = 0.75, 
             cmap = ListedColormap(('yellow', 'white', 'aquamarine'))) 
  
plt.xlim(X1.min(), X1.max()) 
plt.ylim(X2.min(), X2.max()) 
  
for i, j in enumerate(np.unique(y_set)): 
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j) 
  
plt.title('Logistic Regression (Training set)') 
plt.xlabel('PC1') # for Xlabel 
plt.ylabel('PC2') # for Ylabel 
plt.legend() # to show legend 
  
# show scatter plot 
plt.show() 
```

```python
# Visualising the Test set results through scatter plot 
from matplotlib.colors import ListedColormap 
  
X_set, y_set = X_test, y_test 
  
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, 
                     stop = X_set[:, 0].max() + 1, step = 0.01), 
                     np.arange(start = X_set[:, 1].min() - 1, 
                     stop = X_set[:, 1].max() + 1, step = 0.01)) 
  
plt.contourf(X1, X2,lg_model.predict(np.array([X1.ravel(), 
             X2.ravel()]).T).reshape(X1.shape), alpha = 0.75, 
             cmap = ListedColormap(('yellow', 'white', 'aquamarine')))  
  
plt.xlim(X1.min(), X1.max()) 
plt.ylim(X2.min(), X2.max()) 
  
for i, j in enumerate(np.unique(y_set)): 
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], 
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j) 
  
# title for scatter plot 
plt.title('Logistic Regression (Test set)')  
plt.xlabel('PC1') # for Xlabel 
plt.ylabel('PC2') # for Ylabel 
plt.legend() 
  
# show scatter plot 
plt.show() 
```
